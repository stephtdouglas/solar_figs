# Now loop over the whole month and retrieve images

```python
base_url = "https://sdo.gsfc.nasa.gov/assets/img/browse/2011/11/"
```

```python
response = requests.get(base_url)
if response.ok:
    response_text = response.text
else:
    response.raise_for_status()
soup0 = BeautifulSoup(response_text, 'html.parser')
img_dirs = soup0.find_all("a",href=re.compile("/"))

for j, dayfolder in enumerate(img_dirs):
    if "assets" in dayfolder["href"]:
        continue
    day_url = base_url+dayfolder["href"]
    print(day_url)
    
    response = requests.get(day_url)
    if response.ok:
        response_text = response.text
    else:
        response.raise_for_status()
        break
    soup = BeautifulSoup(response_text, 'html.parser')
    hmi_images = soup.find_all("a",href=re.compile("5241_2048_HMII"))
    print(hmi_images)
    
    nimg = len(hmi_images)
    for i in range(nimg):
        img_filename = hmi_images[i]["href"]
        print(img_filename)
        if os.path.exists(save_path+img_filename)==False:
            filestuff = urlretrieve(day_url+img_filename,save_path+img_filename)
            time.sleep(2)
```
